```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
```

```python
x_point=np.array([2,5])
y_point=np.array([0,250])
plt.plot(x_point,y_point)
```

```python
x_point=np.array([1,8])
y_point=np.array([3,10])
plt.plot(x_point,y_point, "o")
```

```python
x=np.array([1,3,5,7,9,0])
y=np.array([10,20,30,40,50,60])
plt.plot(x,y, 'o:g' , ms = 10 ,  mec = 'r' , linewidth=2)
plt.xlabel("Age")
plt.ylabel("Salary")
plt.title("this is my first plot")
plt.rcParams["font.family"]="serif"
plt.rcParams["font.size"]=12
```

```python
plt.scatter(x,y, marker="o",color= "g", cmap="plasma")

print (plt.colormaps())
```

```python
plt.pie(x,autopct="%1.2f%%")
```
